```python
fig,ax = plt.subplots(2,2,figsize=[10,8])

#fig[0,0]
ax[0,0].plot([i for i in range(10)],linestyle='--',color='grey')
ax[0,0].set_title('Line Plot',fontsize=12)
# fig [0,1]
ax[0,1].hist([random.randrange(0,10) for i in range (50)])
ax[0,1].set_title('Histogram')

ax[1,0].scatter([i for i in range(10)],[i for i in range(10)])

# Scatter plot puntos 1
ax[1,1].scatter([random.uniform(1,0) for i in range(100)],[random.uniform(1,0) for i in range(100)],alpha=0.3,color='red')

# Scatter plot puntos 2
ax[1,1].scatter([random.uniform(1,0) for i in range(100)],[random.uniform(1,0) for i in range(100)],alpha=0.4,color='yellow')

# Scatter plot puntos 3
ax[1,1].scatter([random.uniform(1,0) for i in range(100)],[random.uniform(1,0) for i in range(100)],alpha=0.3,color='g')
ax[1,1].set_title('Scatter Plor with 3 diff colors')
plt.show()
```

## Step 6
* Make a new figure with 4 subplots (2x2)
* Make a chart of Open, High, Low and Close in each axis.

```python
fig, ax = plt.subplots(2,2,figsize=[10,10])
# ax[0,0]
ax[0,0].set_title('Open',fontsize=13)
ax[0,0].plot(data['Open'],color='m')
# ax[1,0]
ax[1,0].set_title('Close',fontsize=13)
ax[1,0].plot(data['Close'],color='red')
# ax[0,1]
ax[0,1].set_title('Low',fontsize=13)
ax[0,1].plot(data['Low'],color='green')
# ax[1,1]
ax[1,1].set_title('High',fontsize=13)
ax[1,1].plot(data['High'],color='blue')

plt.show()
```
