```python
import pandas_datareader as pdr
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### Step 2
- Make a datetime object start for 2020-01-01 (Jan. 1st, 2020)
- Make a datetime object end for 2020-12-31 (Dec. 31st, 2020)
- Read data from ticker NFLX from Yahoo! Finance with pandas datareader
- Take head on the dataframe

```python
start = dt.datetime(2020,1,1)
end = dt.datetime(2020,12,31)

nflx = pdr.get_data_yahoo('NFLX',start,end)
nflx.head()
```

### Step 3
- Make a figure with axis from matplotlib
- Make a plot of closing pring (Close)

```python
fig, ax = plt.subplots(figsize=[15,5])
ax.plot(nflx['Adj Close'])
ax.set_title('NFLX - Stock',fontsize=16)
ax.set_xlabel('Date',fontsize=12)
ax.set_ylabel('Price',fontsize=12)
plt.show()
```

### Step 4
- Read data from 4 tickers: "NVDA", "HP", "FB", "GOOG"
- Use the same start and end date
- Also use Yahoo! Finance.
- Take head of the data

```python
tickers=["NVDA", "HP", "FB", "GOOG"]
data = pdr.get_data_yahoo(tickers,start,end)
data.head()
```

### Step 5
- Make 4 plots in 1 figure with the closing price (Close) of the 4 tickers from last step

```python
fig, ax = plt.subplots(figsize=[15,5])
ax.plot(data['Adj Close'],label=tickers)
ax.set_title('Tickers',fontsize=16)
ax.set_xlabel('Date',fontsize=12)
ax.set_ylabel('Price',fontsize=12)
plt.legend()
plt.show()
```

```python
tickers=["NVDA", "HP", "FB", "GOOG"]
def plot(df,nrows=2,ncols=2):
    fig,ax = plt.subplots(nrows,ncols,figsize=[10,10],sharex=True, sharey=False)
    for i in range(nrows):
        for j in range(ncols):
            for ticker in tickers:
                frame = df['Adj Close'][ticker]
                name = frame.name
                ax[i,j].plot(frame)
                ax[i,j].set_title(name)
                break
                plt.tight_layout()
```
